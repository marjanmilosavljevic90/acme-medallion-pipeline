# Set Silver tables

- Read Silver tables from silver_config.py and passes them as a list to the downstream
- For each task which will call 04_run_silver once per table in parallel

```python
from utils.logger import get_logger
from etl_config.silver_config import SILVER_CONFIG

logger = get_logger("set_silver_tables")
```

```python
silver_tables = list(SILVER_CONFIG.keys())
logger.info(f"Silver tables to process: {silver_tables}")

dbutils.jobs.taskValues.set(
    key="silver_tables",
    value=silver_tables
)

logger.info(f"Task value 'silver_tables' set with {len(silver_tables)} tables.")
            
```
